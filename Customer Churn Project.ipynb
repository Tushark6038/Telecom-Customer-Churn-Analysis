### Replacing Blanks with 0 as Tenure is Zero

```python
df["TotalCharges"] = df["TotalCharges"].replace(" ","0")
df["TotalCharges"] = df["TotalCharges"].astype("float")
```

#### Changing the 0 & 1 of the SeniorCitizen to Yes & No

```python
def conv(value):
    if(value == 1): return 'Yes'
    else: return 'No'


df["SeniorCitizen"] = df["SeniorCitizen"].apply(conv)
```

### Till here we have completed Descriptive Analysis.

```python
ax = sns.countplot(x='Churn',data=df)
plt.title("Count of Customers by Churn")
ax.bar_label( ax.containers[0] )
plt.show()
```

### Out of 7043 customers, only 26.5% (1869) have churned, while 73.5% (5174) have stayed.
### This shows the company retains a majority of its customers, but the churn segment is still large enough to analyze and reduce.

```python
gb = df.groupby('Churn').agg({'Churn': "count"})

plt.figure(figsize=(4,4))
plt.pie(gb['Churn'],labels=gb.index, autopct="%1.2f%%")
plt.title("Percentage of Churned Customers")
plt.show()
```

### The pie chart confirms that 26.54% of customers have churned, while 73.46% have stayed.
### Though most customers are retained, over one-fourth churn rate is significant and needs attention.

```python
plt.figure(figsize=(4,4))
sns.countplot( x='gender',data=df, hue='Churn')
plt.title("Churn by Gender")
```

### Churn rates for males and females are nearly identical, indicating that gender has no significant impact on customer churn.
### Both genders have a similar ratio of staying vs. leaving the service.

```python
plt.figure(figsize=(4,4))
sns.countplot( x='SeniorCitizen',data=df, hue='Churn')
plt.title("Churn by SeniorCitizen")
```

### Senior citizens have a higher churn rate compared to non-senior customers, even though they are fewer in number.
### This suggests that age may be a factor in customer retention and seniors might need better support or offers.

```python
plt.figure(figsize = (9,4))
sns.histplot(x = "tenure", data = df, bins = 72, hue = "Churn")
plt.show()
```

### Customers with low tenure (new users) have a much higher churn rate, while those with longer tenure are more likely to stay.
### This shows that improving the early customer experience is crucial to reducing churn.

```python
plt.figure(figsize = (4,4))
ax = sns.countplot(x = "Contract", data = df, hue = "Churn")
ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Contract")
plt.show()
```

### Customers on month-to-month contracts have the highest churn rate, while those on one-year or two-year contracts churn significantly less.
### This shows that longer-term contracts improve customer retention effectively.

```python
df.columns.values
```

```python
cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

fig, axes = plt.subplots(3, 3, figsize=(18,18))
for ax, col in zip(axes.flatten(), cols):
    sns.countplot(data=df, x=col, ax=ax, hue='Churn')
    ax.set_title(col)
    ax.set_xlabel('')
    ax.set_ylabel('Count')
    ax.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

## Insights Gained :
### Customers without online security, backup, tech support, or device protection tend to churn more, showing these services improve retention.
### Fiber optic users show a noticeably higher churn rate than DSL or no internet users.
### People using streaming services (TV & movies) also churn more, likely due to short-term or entertainment-based usage.
### Overall, lack of value-added services is a strong churn indicator.

```python
plt.figure(figsize = (10,4))
ax = sns.countplot(x = "PaymentMethod", data = df, hue = "Churn")
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.title("Churned Customers by Payment Method")
plt.xticks(rotation = 45)
plt.show()
```

### Customers using electronic check as a payment method have the highest churn rate, while those using automatic payments (bank/credit card) churn the least.
### This suggests that auto-payment users are more loyal and stable, possibly due to convenience or commitment.
